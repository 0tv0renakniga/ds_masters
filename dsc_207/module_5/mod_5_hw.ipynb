#### 3.1 Plot histogram of AirBnB prices in San Diego.  [1pt]

1.   This histogram should have price ranges on the x-axis and counts on the y-axis.
2.   Choose an appropriate color and edge-color combination.
3.   Choose an appropriate value for bin. Start with a reasonable number of bins and adjust based on the visual appearance of the histogram. Too few bins may obscure important details, while too many bins can make the distribution appear noisy.
4.   Provide axis labels
5.   Provide a chart title/caption

image.png

```python
def calc_bins(price_range,samples,stdev):
  '''
  calc optimal number of histogram bins based on Scott 1979
  https://academic.oup.com/biomet/article-abstract/66/3/605/232642
  bins =R(n^(1/3))/(3.49Ïƒ)
  R: range of data
  n: sample size
  Ïƒ: standard deviation of data
  '''
  bins = price_range*(samples**(1/3))/(3.49*stdev)
  return(int(bins))

# calculate price range, sample size, and standard deviation
price_range = (listings_df.price.max() -listings_df.price.min())
samples = listings_df.price.shape[0]
stdev = listings_df.price.std()
# calculate number of bins to use for histogram
bin_int = calc_bins(price_range,samples,stdev)

listings_df['price'].hist(bins=162,color='blue', edgecolor='black')
plt.title('histogram of AirBnB prices in San Diego')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```
