### 4.2 Print the number of books that have an average rating within 2 standard deviations from mean average rating of the entire dataset.
[Standard deviation](https://en.wikipedia.org/wiki/Standard_deviation) tells you how spread out the data is. It is a measure of how far each observed value is from the mean. In any distribution, about 95% of values will be within 2 standard deviations of the mean.
Hence, for the given task, we want to find the number of books who have an average rating from `mean-2*std_dev` to `mean+2*std_dev`. [1pt]


```python
# WRITE YOUR CODE HERE
books_mean = books.average_rating.mean()
books_std = books.average_rating.std()
books_two_std = books[(books['average_rating'] > books_mean - 2*books_std) & (books['average_rating'] < books_mean + 2*books_std)]
books_two_std.shape[0]
```

### 4.3 Find books with language code as 'ger' and whose publication date is between 2001 and 2005. (Optional)
(Hint - How do we get the year from the publication date in its current form? Can we use [pd.to_datetime](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) to extract what we need for this task?)

```python
# WRITE YOUR CODE HERE
# get rid of bad dates.. set bad dates to na
books['publication_date'] = pd.to_datetime(books['publication_date'], errors='coerce')
# drop na values and update books data frame
books.dropna(subset=['publication_date'],inplace=True)
# create new df for german books between the years 2001 and 2005
ger_books = books[(books['language_code'] == 'ger') & (books['publication_date'].dt.year >= 2001) & (books['publication_date'].dt.year <= 2005)]
ger_books.head()
```
